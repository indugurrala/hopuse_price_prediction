```python
data=pd.read_csv('Raw_Housing_data.csv')
```

```python
data.info()
```

```python
data.dropna(inplace=True, axis=0,subset=['Sale Price'])
```

```python
data.info()
```

```python
numerical_columns=["No of Bathrooms","Flat Area (in Sqft)","Lot Area (in Sqft)"
                  
                  ,"Area of the House from Basement (in Sqft)", "Latitude",
                   "Longitude","Living Area after Renovation (in Sqft)"]
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='median')
data[numerical_columns]=imputer.fit_transform(data[numerical_columns])
```

```python
data.info()
```

```python
c=['Lot Area (in sqft)','Flat Area (in sqft)','Area of the House from Basement (in sqft)','Living Area after Renovation (in sqft)']
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
data[c]=imputer.fit_transform(data[c])
```

```python
columns=data["Zipcode"].values.reshape(-1,1)
columns.shape
```

```python
column=data["Zipcode"].values.reshape(-1,1)
imputer2=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
data["Zipcode"]=imputer2.fit_transform(column)
```

```python
data.describe()
```

```python
data.info()
```

```python
data['Zipcode']=data['Zipcode'].astype('object')
```

```python
data.info()
```

```python
data['No of Times Visited'].unique()
```

```python
mapping={
    'None' : "0",
    'Once'  :"1",
    'Twice' :"2",
    'Thrice' :"3",
    'Four':"4"
}
    
```

```python
data['No of Times Visited']=data['No of Times Visited'].map(mapping)
```

```python
data['No of Times Visited'].unique()
```

```python
data['Ever Renovated']=np.where(data['Renovated Year']==0,'No' ,'Yes')
```

```python
data.head(10)
```

```python
data['Purchase year']=pd.DatetimeIndex(data['Date House was Sold']).year
```

```python
data['Year since Renovated']=np.where(data['Ever Renovated']=='Yes',abs(data['Purchase year']-data['Renovated Year']),0)
```

```python
data.head(10)
```

```python
data.drop(columns=['Date House was Sold','Renovated Year','Ever Renovated','Purchase year'],inplace=True,axis=1)
```

```python
data.describe()
```

```python
data.to_csv("transformed_data",index=False)
```
